```python
import numpy, random, math
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import random
```

```python
from svm import SupportVectorMachines
```

```python
numpy.random.seed(100)

def circular_data():
    classA = numpy.concatenate((
        numpy.random.randn(5, 2) * 0.3 + [1.5, 0.5],
        numpy.random.randn(5, 2) * 0.3 + [-1.5, 0.5],
        numpy.random.randn(5, 2) * 0.3 + [1.5, -.5],
        numpy.random.randn(5, 2) * 0.3 + [-1.5, -.5],
        ))
    classB=numpy.random.randn(20, 2) * .3 

    inputs=numpy.concatenate((classA, classB))
    targets=numpy.concatenate((numpy.ones(classA.shape[0]), -numpy.ones(classB.shape[0])))

    N = inputs.shape[0] # Number of rows (samples)

    permute = list(range(N))
    random.shuffle(permute)
    inputs = inputs[permute,:]
    targets = targets[permute]


    plt.plot([p[0] for p in classA], [p[1] for p in classA], 'b.')
    plt.plot([p[0] for p in classB], [p[1] for p in classB], 'r.')
    plt.title('Circular data')
    plt.show()


    return inputs, targets, classA, classB


def linear_sep_data():
    classA = numpy.concatenate((
        numpy.random.randn(20, 2) * 0.3 + [1.5, 0.5],
        ))
    classB=numpy.random.randn(20, 2) * .3 + [0.0, -0.5]

    inputs=numpy.concatenate((classA, classB))
    targets=numpy.concatenate((numpy.ones(classA.shape[0]), -numpy.ones(classB.shape[0])))

    N = inputs.shape[0] # Number of rows (samples)

    permute = list(range(N))
    random.shuffle(permute)
    inputs = inputs[permute,:]
    targets = targets[permute]

    plt.plot([p[0] for p in classA], [p[1] for p in classA], 'b.')
    plt.plot([p[0] for p in classB], [p[1] for p in classB], 'r.')
    plt.title('Linearly seperable data')

    plt.show()

    return inputs, targets, classA, classB


def overlapping_data():
    classA = numpy.random.randn(100, 2) *.3
    classB = numpy.random.randn(100, 2) *.3

    inputs=numpy.concatenate((classA, classB))
    targets=numpy.concatenate((numpy.ones(classA.shape[0]), -numpy.ones(classB.shape[0])))

    N = inputs.shape[0] # Number of rows (samples)

    permute = list(range(N))
    random.shuffle(permute)
    inputs = inputs[permute,:]
    targets = targets[permute]


    plt.plot([p[0] for p in classA], [p[1] for p in classA], 'b.')
    plt.plot([p[0] for p in classB], [p[1] for p in classB], 'r.')
    plt.title('Overlapping data')
    plt.show()
    return inputs, targets, classA, classB



def polynomial_data():

    classA = numpy.concatenate((
    numpy.random.randn(10, 2) * 0.3 + [1.5, 0.5],
    numpy.random.randn(5, 2) * 0.3 + [-1.5, 0.5],
    numpy.random.randn(5, 2) * 0.3 + [0.0, -2.0],
    ))

    classB = numpy.concatenate((
    numpy.random.randn(10, 2) * 0.3 + [0.0, -0.5],
    numpy.random.randn(10, 2) * 0.3 + [2.0, -0.5],

    ))

    inputs=numpy.concatenate((classA, classB))
    targets=numpy.concatenate((numpy.ones(classA.shape[0]), -numpy.ones(classB.shape[0])))

    N = inputs.shape[0] # Number of rows (samples)

    permute = list(range(N))
    random.shuffle(permute)
    inputs = inputs[permute,:]
    targets = targets[permute]


    plt.plot([p[0] for p in classA], [p[1] for p in classA], 'b.')
    plt.plot([p[0] for p in classB], [p[1] for p in classB], 'r.')
    plt.title('poly data')
    plt.show()

    return inputs, targets, classA, classB



```

```python
print()
circ_data = circular_data()
lin_data = linear_sep_data()
overlap_data = overlapping_data()
poly_data = polynomial_data()
```

```python
svm = SupportVectorMachines(20, data = lin_data)
svm.runMinimize("linear")  
svm.extract_nonzeroes()   
svm.plot()
```

```python
svm = SupportVectorMachines(20, data = circ_data)
svm.runMinimize("polynomial") 
svm.extract_nonzeroes()   
svm.plot()
```

```python
svm = SupportVectorMachines(20, data = circ_data, C = 100, sigma = 0.1)
svm.runMinimize("rbf") 
svm.extract_nonzeroes()   
svm.plot()
```
